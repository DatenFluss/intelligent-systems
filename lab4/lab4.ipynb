## Task 1 - write out your variant equations in LaTex

\begin{align*}
-10.2x_1 + 9.3x_2 &= 73 \\
5.3x_1 + 15.8x_2 &= 173 \\
17.2x_1 + 10.3x_2 &= 159
\end{align*}

## Task 2 - Using python Math Plot 
plot all functions

```python
import numpy as np
import matplotlib.pyplot as plt

def eq1(x):
    return (73 + 10.2 * x) / 9.3

def eq2(x):
    return (173 - 5.3 * x) / 15.8

def eq3(x):
    return (159 - 17.2 * x) / 10.3

x = np.linspace(-20, 20, 400)
plt.figure(figsize=(12, 8))

plt.plot(x, eq1(x), 'r:', label='$-10.2x_1 + 9.3x_2 = 73$')
plt.plot(x, eq2(x), 'g--', label='$5.3x_1 + 15.8x_2 = 173$')
plt.plot(x, eq3(x), 'b-.', label='$17.2x_1 + 10.3x_2 = 159$')

plt.title('Graphs of the Linear Equations')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(True)
plt.show()
```

## Task 2 - Using python Math Plot 
plot all functions separately

```python
x_range = np.linspace(-10, 10, 400)

plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.plot(x_range, eq1(x_range), 'r:', label='$-10.2x_1 + 9.3x_2 = 73$')
plt.title('Equation 1')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x_range, eq2(x_range), 'g--', label='$5.3x_1 + 15.8x_2 = 173$')
plt.title('Equation 2')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x_range, eq3(x_range), 'b-.', label='$17.2x_1 + 10.3x_2 = 159$')
plt.title('Equation 3')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def find_intersection(eq1, eq2, x_values):
    eq1_y_values = eq1(x_values)
    eq2_y_values = eq2(x_values)
    intersections = []
    for i in range(len(x_values) - 1):
        if (eq1_y_values[i] - eq2_y_values[i]) * (eq1_y_values[i + 1] - eq2_y_values[i + 1]) < 0:
            x_inter = x_values[i] - (x_values[i + 1] - x_values[i]) * (eq1_y_values[i] - eq2_y_values[i]) / \
                      (eq1_y_values[i + 1] - eq1_y_values[i] - (eq2_y_values[i + 1] - eq2_y_values[i]))
            y_inter = eq1(x_inter)
            intersections.append((x_inter, y_inter))
    return intersections

intersections = []
intersections += find_intersection(eq1, eq2, x)
intersections += find_intersection(eq1, eq3, x)
intersections += find_intersection(eq2, eq3, x)

intersections.sort()

intersections_np = np.array(intersections)

plt.figure(figsize=(12, 8))
plt.plot(x, eq1(x), 'r:', label='$-10.2x_1 + 9.3x_2 = 73$')
plt.plot(x, eq2(x), 'g--', label='$5.3x_1 + 15.8x_2 = 173$')
plt.plot(x, eq3(x), 'b-.', label='$17.2x_1 + 10.3x_2 = 159$')

if len(intersections_np) > 0:
    plt.fill(intersections_np[:, 0], intersections_np[:, 1], 'grey', alpha=0.5)

plt.title('Intersection of Three Lines')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(True)
plt.show()
```
